```python
from random import *
from tkinter import *

num = "0123456789"
alphanum = "abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
spcharalphanum = "abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ+-@#(){}%&"

def generate_password():
    selected = Typechoice.get()
    
    if selected =="":
        result.delete(0.0,END)
        result.insert(END,"\n Please select a type of password from the available options")
        
    pass_length = int(password_length.get())
    random_password = []
    for i in range(pass_length):
        random_password.append(choice(selected))
        
    outcome = "".join(random_password)
    
    outcome_presented = "\n *** The generated password is : "+outcome+" ***\n"
    
    result.delete(0.0,END)
    result.insert(END,outcome_presented)
    
    
window = Tk()
window.geometry("800x500") #the size of the window
window.title("Password Generator")
window.iconbitmap(r"/Users/emmanuel_bamidele/Desktop/Bamspace/Data_Science_Tutorial/Python projects/Codes/generator.ico") #to add an icon - convert with convertico

Programname = Label(window, font=('Consolas',30,'bold'), text="Password Generator", fg="blue")

Programname.grid(row=1, column=6, padx=250, pady=30) #method one to show where to place it

SelectType = Label(window, font=('Consolas',20,'italic'), text="Choose the type of your password", fg="blue")
SelectType.place(relx=.03, rely=.23)  #second method of placing

Typechoice = StringVar()
NumberChoice = Radiobutton(window, font=('corbel', 10, 'italic'), text="Numeric", variable = Typechoice, value=num)
NumberChoice.place(relx=.03, rely=.33)  

AlphanumericChoice = Radiobutton(window, font=('corbel', 10, 'italic'), text="Alphanumeric", variable = Typechoice, value=alphanum)
AlphanumericChoice.place(relx=.03, rely=.37)  

CharAlphanumericChoice = Radiobutton(window, font=('corbel', 10, 'italic'), text="Special Character, Aplanumeric", variable = Typechoice, value=spcharalphanum)
CharAlphanumericChoice.place(relx=.03, rely=.41)

length = Label(window, text="Length", font=('Copperplate Gothic',15, 'bold'))
length.place(relx=.62, rely=.35)

password_length = Spinbox(window, from_=8, to=100)
password_length.place(relx=.71, rely=.35)
password_length.config(state="readonly") #to avoid typing into the spinbox

Generate = Button(window, text="Generate Password", command=generate_password) #link this to the function
Generate.place(relx=.41, rely=.61)

result = Text(window, height=6, width=70, wrap=WORD)
result.place(relx=.21, rely=.68)

window.mainloop() #to keep the window open continuously
```
