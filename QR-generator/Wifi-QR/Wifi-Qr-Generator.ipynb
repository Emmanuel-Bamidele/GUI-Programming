```python
import wifi_qrcode_generator as qr
from tkinter import *
from tkinter import filedialog
import matplotlib.pyplot as plt


def qr_generator():
    wifi_name = str(wifi.get())
    wifi_ps = str(wifi_p.get())
    fname = str(filename.get())
    
    qr_out = qr.wifi_qrcode(wifi_name, False, 'WPA', wifi_ps)
    qr_out.save(fname + ".png")
    #fname = outpath + "/" + fname
    
    result.delete(0.0, END)
    result.insert(END, "Successful...")


window = Tk()
window.geometry("250x300")  # the size of the window
window.title("Â© Bamidele 2021")
window.iconbitmap(
    r"/Users/emmanuel_bamidele/Desktop/Bamspace/Data_Science_Tutorial/Python projects/Codes/QR-Generator/qr-icon.ico")

Programname = Label(window, font=('Verdana', 25, 'bold'), text="QR-WIFI", fg="blue")
Programname.place(relx=0.2, rely=0.02)

wifi_label = Label(window, text="Wifi Name", font=('Copperplate Gothic', 15, 'bold'))
wifi_label.place(relx=.3, rely=.15)

wifi = Entry(window)
wifi.place(relx=.08, rely=.25)

wifi_p_label = Label(window, text="Password", font=('Copperplate Gothic', 15, 'bold'))
wifi_p_label.place(relx=.3, rely=.38)

wifi_p = Entry(window, show="*")
wifi_p.place(relx=.08, rely=.48)

file_label = Label(window, text="File Name", font=('Copperplate Gothic', 15, 'bold'))
file_label.place(relx=.3, rely=.60)

filename = Entry(window)
filename.place(relx=.08, rely=.70)

Generate = Button(window, text="Generate", command=qr_generator)
Generate.place(relx=.32, rely=.80)

result = Text(window, height=1, width=20, wrap=WORD, fg="red")
result.insert(END, "Enter Inputs...")
result.place(relx=.2, rely=.90)

#out_path = filedialog.askdirectory()

window.mainloop()
```
